# **Vectorized String operations**
Vectorized operations are operations that are applied to entire column rather than individual rows. This approach is much faster than looping through elements one-by-one, as it leverages low-level optimizations

# Problem with vectorized operations

Python will not be able to handle if there is 'None' values

EG:

```python
s=['cat','dog',None,'rat']
[i.startswith('c') for i in s]
```

# **How pandas solved above occured issue?**
EG:

```python
s=pd.Series(['cat','dog',None,'rat'])
s.str.startswith('c')
```

```python
import numpy as np
import pandas as pd
```

```python
#import titanic dataset
df=pd.read_csv('/content/titanic.csv')
df.head()
```

```python
df['Name']
```

```python
#common functions
#lower/uppper/captialize/title
```

```python
#lower
df['Name'].str.lower()
```

```python
#upper()
df['Name'].str.upper()
```

```python
#captialize-First letter of string to capital lettter
df['Name'].str.capitalize()
```

```python
#title()-every word of string to captial
df['Name'].str.title()
```

```python
#len
df['Name'][df['Name'].str.len()==82].values[0]
```

```python
#strip()-to strip necesaary leading or trailing values in our string
df['Name'].str.strip()
```

```python
# split->get
df['lastname']=df['Name'].str.split(',').str.get(0)
```

```python
df[['title','firstname']]=df['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
df.head()
```

```python
#replace
df['title']=df['title'].str.replace('Ms.','Miss.')
df['title']=df['title'].str.replace('Mlle.','Miss.')
df['title']=df['title'].str.replace('Mme.','Mrs.')
df['title'].value_counts()
```

```python
#filtering
#startswith()
#endswith()
#isdigit()
df[df['firstname'].str.isdigit()]
```

```python
#apply regs
#search john->both case
df[df['firstname'].str.contains('john',case=False)]
```

```python
#Find the lastnames which starts and ends with vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
#slicing
df['Name'].str[0:7]
```
