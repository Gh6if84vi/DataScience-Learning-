# apply(Series+DF)

```python
temp=pd.Series([1,2,3,4,5])
def square(x):
  return x**2
temp.apply(square)
```

```python
points_df=pd.DataFrame({
    '1st point':[(3,4),(-6,5),[0,0],[-10,1],[4,5]],
    '2nd point':[(4,3),(5,-6),[0,1],[1,-10],[5,4]]
})
points_df
```

```python
def euclidiean(row):
  pt_a=row['1st point']
  pt_b=row['2nd point']
  return ((pt_a[0]-pt_b[0])**2+(pt_a[1]-pt_b[1])**2)**0.5
```

```python
points_df["Distance"]=points_df.apply(euclidiean,axis=1)
points_df
```
