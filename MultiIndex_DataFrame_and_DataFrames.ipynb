# **Stacking and Unstacking**

# Unstack() :
**unstack() typically converts multiindex objext into a DF by removing inner rows and making them as columns**

# Stack() :
**stack() typically converts DF to multiIndex object by removing inner columns and making them as rows or indexes**



```python
#1
branch_df1
```

```python
#2
branch_df1.unstack()
```

```python
#3
branch_df1.unstack().unstack().unstack()
```

```python
#Now see 1 and 2 to know the differnec after unstack() on multiindex object
#normal bf1 was multiindex on indexes which became multiindex on columns after unstacking
```

```python
branch_df1.unstack().unstack().unstack().stack()
```

```python
branch_df1.unstack().unstack().unstack().stack().unstack()
```

# Long Vs Wide Data

**Wide format** is where we have a single row for every data poiny with multiple columns to hold the values of various attributes

**Long format** is where.for each data point we have as many rows as the no.of attributes and each row contains the value of a particular attribute for a given data point

# melt()-wide to long

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]})
```

```python
pd.DataFrame({'cse':[120],'ece':[100],'mech':[50]}).melt(var_name='branch',value_name='no.of_students')
```

```python
df=pd.DataFrame({
    'branch':['cse','ece','mech'],
    '2019':[120,110,50],
    '2020':[120,100,40],
    '2021':[150,80,30]
})
df
```

```python
df.melt(id_vars=['branch'],var_name='year',value_name='no.of_students')
```
