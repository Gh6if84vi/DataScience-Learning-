# 11.Encoding parameter

# 1. UTF-8
UTF-8 is the most widely used encoding format for text data. It supports all characters in the Unicode standard and is compatible with ASCII. To specify UTF-8 encoding in Pandas, use the encoding='utf-8' parameter.

import pandas as pd

df = pd.read_csv('file.csv', encoding='utf-8')

# 2. ISO-8859-1
ISO-8859-1, also known as Latin-1, is another popular encoding format for text data. It supports most Western European languages and is compatible with ASCII. To specify ISO-8859-1 encoding in Pandas, use the encoding='iso-8859-1' parameter.

import pandas as pd

df = pd.read_csv('file.csv', encoding='iso-8859-1')

# 3. ASCII
ASCII is a 7-bit encoding format that represents only the basic English alphabet, digits, and punctuation marks. To specify ASCII encoding in Pandas, use the encoding='ascii' parameter.

import pandas as pd

df = pd.read_csv('file.csv', encoding='ascii')

date-object,is what actually present in the dataset.So when we try to 
access year or month,we get error as it is string.So to tackle this kind 
of errors,we use "parse_dates" parameter.

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/Gh6if84vi/100-days-of-machine-learning/main/day15%20-%20working%20with%20csv%20files/IPL%20Matches%202008-2020.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data4 = StringIO(req.text)
df=pd.read_csv(data4,parse_dates=['date'])
df.info()

#now see the output at date,you will see the difference
```

# 15.Convertors

description relates to the converters parameter in functions like pandas.read_csv(), 
which allows you to specify custom conversion functions for certain columns when reading data from a CSV file.

converters : dict, default None
It accepts a dictionary where:
Keys: Can be column names (strings) or column indices (integers).
Values: Are functions that take a single argument (the cell value) and return the converted value.

This is useful when you need to preprocess certain columns during data loading, 
such as transforming strings to numeric types, handling missing values, or any custom transformation.

```python
def convert_price(val):
    return float(val.strip('$'))
convert_price('$2345')
#pd.read_csv('data.csv',convertors={'price':convert_price})
```
